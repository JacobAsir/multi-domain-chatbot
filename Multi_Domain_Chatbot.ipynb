```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B")

# Add a padding token to the tokenizer if it doesn't have one
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})
```

```python
from transformers import LlamaForCausalLM

# Load the base model
model = LlamaForCausalLM.from_pretrained('meta-llama/Llama-3.2-1B')

# Add the pad token to the model's embedding
model.resize_token_embeddings(len(tokenizer))
```

```python
def preprocess_function(examples):
    # Combine 'domain' and 'question' as strings
    inputs = [f"{domain} | {question}" for domain, question in zip(examples['domain'], examples['question'])]
    targets = [str(answer) for answer in examples['answer']]  # Ensure targets are strings

    # Tokenize inputs
    model_inputs = tokenizer(
        inputs,
        truncation=True,
        padding='max_length',  # Padding to max length
        max_length=64,        # Adjust based on your needs
        return_tensors="pt"
    )

    # Tokenize targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            targets,
            truncation=True,
            padding='max_length',  # Padding to max length
            max_length=64,        # Adjust based on your needs
            return_tensors="pt"
        )

    # Replace pad token id's in labels by -100 to ignore them in loss
    labels["input_ids"][labels["input_ids"] == tokenizer.pad_token_id] = -100

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
train_tokenized = train_dataset.map(preprocess_function, batched=True)
```

```python
from transformers import LlamaForCausalLM
from peft import get_peft_config, get_peft_model, LoraConfig, TaskType

# Define LoRA configuration
peft_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=8,                # Rank of the adaptation
    lora_alpha=32,      # Scaling factor
    lora_dropout=0.1,   # Dropout rate
    target_modules=['q_proj', 'v_proj'],  # Target modules for adaptation, adjust based on architecture
)

# Apply PEFT to the model
model = get_peft_model(model, peft_config)
```

```python
training_args = TrainingArguments(
    output_dir="MultiDomainChatbot",
    learning_rate=2e-5,
    num_train_epochs=3,
    per_device_train_batch_size=1,       # Reduced
    per_device_eval_batch_size=1,        # Reduced
    gradient_accumulation_steps=2,       # Added
    eval_strategy="epoch",
    save_strategy="epoch",
    weight_decay=0.01,
    load_best_model_at_end=True,
    push_to_hub=True,                    # Disabled unless needed
    logging_dir='./logs',
    logging_steps=100,
    report_to="none",                     # Temporarily disable wandb
)
```

```python
trainer = Trainer(
    model = model,
    args = training_args,
    train_dataset = train_tokenized,
    eval_dataset = test_tokenized,
)
```

```python
import torch
torch.cuda.empty_cache()
```

```python
trainer.train()
```
